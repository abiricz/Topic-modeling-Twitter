```python
import gzip
import io
from os import listdir
from os.path import isfile, join
```

### Merging compressed files

```python
def do_merge( filename_in, file_out ):

    with io.TextIOWrapper( io.BufferedReader(gzip.open(filename_in)) ) as file_in:
        for line in file_in: # read one line at a time from a compressed file
            file_out.write( line.encode() ) # write line to file
    file_in.close()
```

### Getting all file names in the directory

```python
dir_path = '/media/abiricz/Elements/twitter/filtered/'
target_path = '/media/abiricz/Elements/twitter/' # save merged file here
dir_files = [f for f in listdir(dir_path)]
```
